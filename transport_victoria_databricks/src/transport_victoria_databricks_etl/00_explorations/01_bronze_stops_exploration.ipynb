```python
%sql
SELECT *
FROM transport_vic_dev.`02_silver`.trip_updates t
LEFT JOIN transport_vic_dev.`02_silver`.stops s
ON t.stop_id = s.stop_id
WHERE t.stop_id = '20043'
ORDER BY t.entity_id, t.trip_id, t.start_date, t.start_time, t.stop_sequence;
```

```python
%sql

WITH ordered AS (
  SELECT *,
    LAG(struct(arrival_time, arrival_delay_sec, departure_time,
               departure_delay_sec, stop_rel))
      OVER (PARTITION BY entity_id, stop_sequence ORDER BY feed_ts) AS prev
  FROM transport_vic_dev.`02_silver`.trip_updates
)
SELECT * EXCEPT(prev) FROM ordered
WHERE prev IS NULL
   OR struct(arrival_time, arrival_delay_sec, departure_time,
             departure_delay_sec, stop_rel) IS DISTINCT FROM prev;
```
